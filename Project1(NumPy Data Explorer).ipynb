```python
import numpy as np
import time

```

```python
# Array Creation
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.arange(1, 21).reshape(4, 5)

print("Array 1:", arr1)
print("Array 2:\n", arr2)

```

```python
# Indexing & Slicing
print("\nFirst Row:", arr2[0])
print("Last Column:", arr2[:, -1])
print("Sub Array:\n", arr2[1:3, 1:4])

```

```python
# Mathematical Operations
print("\nAddition:", arr2 + 10)
print("Multiplication:", arr2 * 2)
```

```python
# Statistical Operations
print("\nMean:", np.mean(arr2))
print("Median:", np.median(arr2))
print("Max:", np.max(arr2))
print("Min:", np.min(arr2))
print("Sum (Axis=0):", np.sum(arr2, axis=0))
```

```python
# Reshaping
reshaped = arr2.reshape(5, 4)
print("\nReshaped Array:\n", reshaped)
```

```python
# Broadcasting
b = np.array([1, 2, 3, 4, 5])
print("\nBroadcasting Result:\n", arr2 + b)
```

```python
# Save & Load
np.save("data.npy", arr2)
loaded = np.load("data.npy")
print("\nLoaded Array:\n", loaded)
```
